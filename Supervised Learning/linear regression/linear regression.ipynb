```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("C://Users//bhavy//OneDrive//Desktop//Machine learning//linear regression//Book1.csv")
df
```

```python
%matplotlib inline
plt.xlabel('Area')
plt.ylabel('Prices')
plt.scatter(df.area, df.price, color='red')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3000]])
```

```python
reg.coef_
#m = slope
```

```python
reg.intercept_
#b = interncept
```

```python
135.78767123*3000+180616.43835616432
#y = dependent variable = prices
#y = m*x + b
#x = area = independent variable
```

```python
%matplotlib inline
plt.xlabel('area', fontsize=20)
plt.ylabel('prices', fontsize=20)
plt.scatter(df.area, df.price, color='red')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```
